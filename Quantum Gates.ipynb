Let's import the necessary modules for making simple quantum gates

```python
from math import pi,sqrt
from qiskit import QuantumCircuit,Aer,execute
from qiskit.tools.visualization import plot_bloch_multivector , plot_histogram 
```

# Pauli's X-gate (X - gate)

```python
x_gate = QuantumCircuit(1)
x_gate.x(0)
x_gate.draw(output = 'mpl')
```

The result of the above gate can be displayed with plot_bloch_multivector()

```python
# %matplotlib inline
import warnings
warnings.filterwarnings('ignore')
sim = Aer.get_backend('statevector_simulator')
state = execute(x_gate,backend = sim).result().get_statevector()
plot_bloch_multivector(state)
```

As expected after applying X-gate to state 0 we get state 1 which is equilvalent to applying rotation by 180 degrees on the bloch sphere.

#  Hadamard Gate


This gate allows us to create a superposition of state 0 and 1 and is implemented as follow

```python
hadamard = QuantumCircuit(1)
hadamard.h(0)
hadamard.draw(output = 'mpl')
```

Let's visualize the hadarmard gate using bloch sphere

```python
sim = Aer.get_backend('statevector_simulator')
state = execute(hadamard,backend = sim).result().get_statevector()
plot_bloch_multivector(state)
```

For better visualization let's use plot_state_qshere

```python
from qiskit.tools.visualization import plot_state_qsphere
sim = Aer.get_backend('statevector_simulator')
state = execute(hadamard,backend = sim).result().get_statevector()
plot_state_qsphere(state)
```
